```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
```

```python
def plot(folders, file, seed, y_label, save_file_name):
    # Load iterations and train-loss data from each folder
    data = []
    for folder in folders:
        iterations = np.load(os.path.join('..', "analysis_results", folder, 'iteration_list.npy'))
        train_loss = np.load(os.path.join('..', "analysis_results", folder, file))
        label = folder.replace(seed, "")  # Remove "seed=0" from the folder name
        label = label.replace("ntrain=", "$N_{train}=$").replace("neval=", "$N_{eval}=$").replace("-", ", ")
        data.append((iterations, train_loss, label))
    
    # Plot the train_losses
    plt.figure(figsize=(10, 6))
    for i, (iterations, train_loss, label) in enumerate(data):
        plt.plot(iterations, train_loss, label=label)

    plt.yscale("log")
    plt.xlabel('Iterations', fontsize=18)  # Increase font size of x-axis label
    plt.ylabel(y_label, fontsize=18)  # Increase font size of y-axis label
    plt.legend(fontsize=14)  # Increase font size of legend text
    plt.tick_params(axis='both', which='major', labelsize=14)  # Increase font size of tick labels
    plt.savefig(os.path.join(os.getcwd(), "figures", save_file_name))
    plt.show()
    
    # plt.title('Losses from Different Folders')
    # plt.grid(True)
    # plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0)) # Set x-axis tick label format to scientific notation
```

```python
# Define the folders
folders = ["ntrain=500-neval=16-seed=0", "ntrain=500-neval=512-seed=0", "ntrain=500-neval=1024-seed=0"]
plot(folders, 'train_loss_list.npy','-seed=0', 'Training Loss', 'ntrain=500_train_losses.pdf')
plot(folders, 'test_loss_list.npy','-seed=0', 'Testing Loss', 'ntrain=500_test_losses.pdf')
```
